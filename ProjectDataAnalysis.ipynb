 # Observations
**1)** Dataset consists of 240 rows, 6 feature and 1 target columns,
- *Absolute Temperature (in K)*    
- *Relative Luminosity (L/Lo)*
- *Relative Radius (R/Ro)*
- *Absolute Magnitude (Mv)*
- *Star Color*
- *Spectral Class*
- *Star Type (Target classes)*
    
Here,
- *Lo = 3.828 x 10^26 Watts (Avg Luminosity of Sun)*
- *Ro = 6.9551 x 10^8 m (Avg Radius of Sun)*
    
**2)** Two categorical features (object type) which will need some kind of encoding,
- *Star Color*
- *Spectral Class*

**3)** It consists of some different features of stars. Information on the star type is given below,       
- **0** → Brown Dwarf       
- **1** → Red Dwarf       
- **2** → White Dwarf      
- **3** → Main Sequence    
- **4** → Supergiants      
- **5** → Hypergiants

```python
# Creating a directory to save visulaizations :-> 
folder_name = 'star_plots'
os.makedirs(folder_name , exist_ok = True)
base_dir = f'{folder_name}/'
```

```python
# Bar Chart to visualize the count of stars with respective type : 
plt.figure(figsize = (4.3, 3.7))
plt.style.use('dark_background')
ax = star_df['Star type'].value_counts().plot(kind='bar' ,color = ['brown' , 'yellow' , 'orange' ,'lightblue' ,  'pink' , 'purple'])
ax.bar_label(ax.containers[0] , color = 'yellow')
plt.title('Visualize Star Count per Star Type')
plt.ylabel('Count of Each Star' , color= 'yellow')
plt.xticks(ticks=[0,1,2,3,4,5] , labels = ['Brown\nDwarf' , 'Red\nDwarf' , 'White\nDwarf'  , 'Main\nSequence' ,'Super\ngiants' ,'Hyper\ngiants'] , rotation = 45)

# Svaing the data into base directory 
plt.savefig(base_dir + 'barplot_star_count.png')
plt.show()
```

<h3>Visualizing some part of data using  SEABORN  Bar Graph</h3>

```python
ax =sns.barplot(x = star_df['Star color'].value_counts().index, 
           y = star_df['Star color'].value_counts()  , legend = False , hue = [0,1,2,3,4])
ax.bar_label(ax.containers[0] , color = 'red' , weight = 'bold')
plt.title('Star Count Chart' , color = 'royalblue'  ,fontsize = 15 )
plt.ylabel('Star Color' , color = 'white' ,fontsize = 13 )
plt.yticks(color = 'tab:pink')
plt.xticks(color = 'white' , rotation = 45 )
#  Saving the image 
plt.savefig(base_dir + 'star_color_viz.png')
plt.show()
```

```python
# Creating function to perfor this task :: 
def linePlotFun(colors , star_df , i):
    plt.subplot(4,1,i+1)
    plt.plot(star_df.iloc[:, 0] , color = colors[i] , )
    plt.title(star_df.columns[i] , color = 'pink')
```

```python
#Line Plot 

colors = ['gold' , 'silver' , 'lime', 'magenta']
plt.figure(figsize= (15,8))
plt.suptitle('Visualizing the distribution of Numeric Features\n' , fontsize = 15)
for i in range(4):
    linePlotFun(colors , star_df  , i )
plt.tight_layout()
plt.show()
```
